```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 150
```

```python
def showImage(img,title="none"):
    if(len(img.shape) > 2):
        plt.imshow( cv.cvtColor(img, cv.COLOR_BGR2RGB) , cmap="gray")
        plt.title(title)
        plt.show()
    else:
        plt.imshow( img , cmap="gray")
        plt.title(title)
        plt.show()
```

```python
img = cv.imread("images/test_overlap1.tif",-1)
img2 = cv.imread("images/test_overlap2.tif",-1)
```

```python
def tileImage(image,kernel_size):
    '''
        Cut the image into tiles of kernel size x and y
        example;
         1280x960 image and kernel size is 5x5
         should return 25 tiles of size 256x192 
    '''
    if(len(img.shape)==2):
        img_height,img_width = img.shape[:2]
        channels = 1
    else:
        img_height,img_width,channels = img.shape[:3]
    tile_height,tile_width = kernel_size
    # split the image dimensions by tile size
    num_x = int(img_width/tile_width)
    num_y = int(img_height/tile_height)
    
    tiles = []
    for y in range(0,img_height,num_y):
        for x in range(0,img_width,num_x):
            # extract tiles from the image
            tiles.append( image[y:y+num_y,x:x+num_x] )
    return tiles
```

```python
def intertwineTiles(tiles1,tiles2,kernel_size,image_size,dtype):
    '''
        Simply take the 2 tile arrays and alternate the tiles
        first array if the counter is divisible by 2
        second array otherwise
        
        then hstack the horizontal tiles into strips
        then vstack the horizontal strips
    '''
    img_height,img_width = image_size
    
    tile_height,tile_width = kernel_size
    length = len(tiles1)
    
    overlay = []
    for i in range(length):
        if(i%2==0):
            overlay.append(tiles1[i])
        else:
            overlay.append(tiles2[i])
    
    horizontal_strips = []
    for i in range(tile_height):
        start_index = tile_width*(i-1)+tile_height
        end_index = tile_width*(i)+tile_height
        #print(start_index,end_index)
        horizontal_strips.append(np.hstack(overlay[start_index:end_index]))
    
    return np.vstack(horizontal_strips)
```

```python
tiles = tileImage(img,(5,5))
tiles2 = tileImage(img2,(5,5))
overlay = intertwineTiles(tiles,tiles2,(5,5),img.shape,img.dtype)
```
